# Do Opposites Attract?
_____

In order to answer the age-old question "do opposites attract?", first attraction needs to be defined as it pertains to the dataset. Seeing as the end result of speed dating is either a matched pair or not, the group decided to define attraction as such (i.e. 'match' = 1). According to the provided documention, a match only occurs when both participants answer 'yes' when asked if they would like to see each other again (i.e. 'dec_ind' = 1 and 'dec_par' = 1). The group assumed that if both individuals want to see each other, then they are attracted to each other.

Since opposite is defined as anything that is not the same or similar, the group decided to use multiple factors to answer our main question. In particular, we investigated and asked questions revolving around: age, career, interests, and intention. Below is how we defined 'opposite' for each:
* age: larger age gaps (opposite) versus smaller age gaps/no age gap 
* career: different career (opposite) versus same career
* interests: different levels of interest in a particular activity (opposite) versus same level of interest
* intention: different personal goals when it comes to participating in speed dating (opposite) versus the same goal 

The source of data is from Kaggle and it is based on a study from Columbia Business School. The dataset includes a CSV file as well as a dataset key explaining the column names.

https://www.kaggle.com/annavictoria/speed-dating-experiment/version/1

http://www.stat.columbia.edu/~gelman/arm/examples/speed.dating/

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns

# Save each file path to a variable
sd_csv = "Resources/speed_dating_data.csv"

# Import the file into two seperate dataframes: one for the individual's data (iid) and the second for the partner's data (pid)
ind_df = pd.read_csv(sd_csv, encoding="ISO-8859-1")
par_df = pd.read_csv(sd_csv, encoding="ISO-8859-1")
```

## Data Exploration & Clean Up
____
### Overall Data Clean Up

The original dataset contains 195 columns, all of which are explained in the accompanying speed dating data key.

Each row of the dataset corresponds to an individual's (iid) speed dating experience with a specific partner (pid), as well as additional information related to that person's views on the experience, their career,  age,  race, interests, intention, etc.

In order to analyze each speed date, the original dataset has been saved into two seperate dataframes, one corresponding to the individual's responses (ind_df), and the other for their partner's responses (par_df). 

Since only certain columns will be analyzed below, copies of both ind_df and par_df are created in order to manipulate those dataframes, clean_ind and clean_par respectively. Next, each dataframe is truncated to only hold columns that are going to be investigated. Once the columns for each dataframe are specified, both dataframes are merged into one (combo_df) using the iid column from clean_ind and the pid column from clean_par since those columns align to the same value. Suffixes are also added in order to distinguish what columns are coming from each dataframe. Furthermore, we only want the rows were pid_ind corresponds to iid_par. 

```python
# Create a copy of the ind_df in order to manipulate the dataframe
clean_ind = ind_df.copy()
# Select only the columns that will be analyzed (i.e. age, race, career, intention, interests)
clean_ind = clean_ind[["iid","pid","match","age","career_c","goal",
                       "sports","tvsports","exercise","dining","museums","art","hiking","gaming","clubbing",
                       "reading","tv","theater","movies","concerts","music","shopping","yoga"]]

# Create a copy of the par_df in order to manipulate the dataframe
clean_par = par_df.copy()
# Select only the columns that will be analyzed (i.e. age, race, career, intention, interests)
clean_par = clean_par[["iid","pid","age","career_c","goal",
                       "sports","tvsports","exercise","dining","museums","art","hiking","gaming","clubbing",
                       "reading","tv","theater","movies","concerts","music","shopping","yoga"]]

# Merge clean_ind and clean_par using the iid column for clean_ind (left_on) and pid column for clean_par (right_on)
# Also add '_ind' as the suffix for data related to iid's responses and add '_par' for data related to pid's responses
combo_df = pd.merge(clean_ind,clean_par, left_on = 'iid',right_on='pid',suffixes=("_ind", "_par"))
# Specify that the pid_ind column corresponds to the idd_par column
combo_df = combo_df[combo_df['pid_ind'] == combo_df['iid_par']]
```

Upon merging the cleaned individual dataframe and the cleaned partner dataframe, it became apparent that the rows were not unique. What this meant for the data was that instead of having one row for each pair of iid and pid, there was a repeat row since iid's and pid's are interchangeable, depending on whose response is being analyzed. For example, the first row in the dataframe is the speed date experience for iid_ind = 1 and pid_ind = 11. However, further along the dataframe, there will be a row with the same information except that now the iid_ind = 11 and pid_ind = 1. Because of this duplication, a for-loop was used to create unique pairings for each date using the zip funtion on the iid_ind column and the iid_par column as well as the set function for each unique pair.

```python
# Create two empty lists
found = []
found_sets = []
# For each individual iid (ind) and each partner iid (par) in the zip function (iterator) of those respective columns
for ind, par in zip(combo_df['iid_ind'], combo_df['iid_par']):
    # If the set of individial iid and partner idd are already in found_sets (i.e. not unique)
    if set([ind, par]) in found_sets:
        # Add False to the found list
        found.append(False)
    else:
        # Add True to the found list
        found.append(True)
        # Add the unique set of individual iid and partner iid to the found_sets list
        found_sets.append(set([ind, par]))

# Isolate combo_df so that all speed dating experiences are unique
combo_df = combo_df[found]

# Print the first five rows
combo_df.head()
```

### Do different interests lead to successful matches?

In the speed dating dataset that was investigated, participants were asked to rank their interest in the following 17 activities, on a scale from 1 to 10.

* sports: Playing sports/ athletics
* tvsports: Watching sports
* exercise: Body building/exercising
* dining: Dining out
* museums: Museums/galleries
* art: Art
* hiking:  Hiking/camping
* gaming: Gaming
* clubbing: Dancing/clubbing
* reading: Reading
* tv: Watching TV
* theater: Theater
* movies: Movies
* concerts: Going to concerts
* music: Music
* shopping: Shopping
* yoga: Yoga/meditation

As one of the ways to prove the null hypothesis that opposites attract, the level of interest in the above activities was compared to determine if a pair who share the same interest levels will less frequently result in a match than those with different interests.

For the initial analysis, a heatmap was created for each of the 17 activities using a for-loop. Using a heatmap allows the viewer to see any overall patterns between two categories of data. As the shade of purple becomes darker, this means that a higher proportion of pairs matched given their respective level of interest in an activity. To prove our hypothesis, we expect to see darker purple squares away from the positive sloping diagonal. If darker squares fall along the positive sloping diagonal, this means that a higher percentage of pairings scored a particular activity the same.

```python
# Create a copy of the combo_interests in order to manipulate the dataframe
activities_df = combo_interests.copy()

# For each column in the specified individual interests columns of the dataframe,...
for column in activities_df[int_columns_ind]:
    # For each activity in the list of 17 activities (defined in Data Clean Up)
    for activity in activities:
        # If the column name begins with the specific activity
        if column.startswith(activity):
            # Create a new dataframe for that activity with only the individual's interest level in the activity, their
            # partner's interest level in the activity and if they matched (match = 1 or match = 0)
            specific_df = activities_df[[f'{activity}_ind',f'{activity}_par','match']]
            # Set all columns to integers since they are all discrete values
            specific_df = specific_df.astype('int')
            # Create a pivot table that averages the matches based on the individual's interest level and their partner's
            specific_df = specific_df.pivot_table('match',f'{activity}_ind',f'{activity}_par',aggfunc=np.average)
            # Create a heatmap for each activity and title it appropriately
            ax = sns.heatmap(specific_df,cmap='BuPu')
            ax.set_title(f"Average Interest Levels in {activity.upper()}")
            # Invert the y-axis so that correlation between interest levels is more readable
            # i.e. if darker shades of purple lie on the positive diagonal slope, then the pair ranked the activity the same
            ax.invert_yaxis()
            # Save as .png
            plt.savefig(f"Visuals/interests/Average Interest Levels in {activity.upper()}.png", transparent=False)
            # Display the heatmap
            plt.show()
```

The above heatmaps imply that:
* there was no activity heatmap that had a clear positive diagonal slope (i.e. that only those with identical interest levels in an activity were successful matches)
* for exercise, hiking, shopping, sports, TV and TV Sports, there was some patterns forming around the positive diagonal slope, but there were plently of dark purple squares elsewhere on the heatmap
* for concerts, dining, gaming, music, reading and theater, there were only one or two dark purple squares (i.e. an identifier of increased match proportion) and they were located away from the positive diagonal slope, meaning that these could be indications that opposites attract

While none of the heatmaps showed a clear positive sloping diagonal (i.e. that likes attract) and had scattered dark purple squares away from said line (i.e. potentially implying that opposites do attract), additional analysis between matched and unmatched pairs had to be conducted. Specifically, if the absolute difference between each participant's and their partner's interest level were calculated, would the results differ in matches and non-matches?

Each plot showed that more matches occured when couples shared similar levels of interest in an activity (i.e. skewed to the right), thereby failing to prove our null hypothesis that opposites attract. However, a similar result was also found for couples who did not match (i.e. similar levels of interest resulted in more non-matches).

```python
# Create a copy of the cleaned_interests in order to manipulate the dataframe
matched_interests = cleaned_interests.copy()
# Only select rows with matches (i.e. match = 1)
matched_interests = matched_interests[matched_interests['match'] == 1]

# Create a copy of the cleaned_interests in order to manipulate the dataframe
unmatched_interests = cleaned_interests.copy()
# Only select rows with non-matches (i.e. match = 0) 
unmatched_interests = unmatched_interests[unmatched_interests['match'] == 0]

# Set variables for the total number of rows in the matched_interests and unmatched_interests dataframes
total_matched = matched_interests['match'].count()
total_unmatched = unmatched_interests['match'].count()
```

```python
# For each activity in the list of 17 activities...
for activity in activities:
    # Use groupby to separate data into fields based on a matched couple's difference in interest level
    m_group = matched_interests.groupby(activity)
    # Use groupby to separate data into fields based on an unmatched couple's difference in interest level
    u_group = unmatched_interests.groupby(activity)
    # Use the count function on the grouped data based on the match column
    m_count = m_group['match'].count()
    u_count = u_group['match'].count()
    
    # Set variables for the subsequent plots
    fig, ax = plt.subplots()
    # Create a bar graph showcasing how many matches had differing interest levels
    # If difference equals zero, this means that they ranked that activity the same
    # If the difference equals nine, this means that they ranked the activity completely opposingly (i.e. a 1 and a 10)
    m_count.plot(kind = 'bar')
    # Label the y-axis accordingly
    plt.ylabel('Matched Pairs')
    # Create a line graph showing how many non-matches had differing interest levels using a secondary y-axis
    # By using a secondary y-axis, the overall shape of the data is visible without worrying about scale
    # This is important since there were significantly more non-matches than matches
    u_count.plot(kind = 'line', marker = 'o', secondary_y = True, color='magenta')
    # Label the secondary y-axis accordingly
    plt.ylabel('Non-matched Pairs',rotation = 90,labelpad=5)
    # Set the x boundary for better visibility
    ax.set_xbound(lower=-0.5,upper=9.5)
    # Title the plot
    ax.set_title(f'Difference in Interest Levels for {activity.upper()}')
    # Save as .png
    plt.savefig(f"Visuals/interests/Difference in Interest Levels for {activity.upper()}.png", transparent=False)
    # Display the plots
    plt.show()
    
    # Create a pandas Dataframe with columns containing for matched and unmatched counts per differing interest level
    interest_df = pd.DataFrame(dict(matched = m_count, unmatched = u_count))
    # For complete analysis and to ensure the same number of degrees of freedom, change NaN to zero 
    # (e.g. largest difference in matched group was 8 while the largest difference in the unmatched group was 9)
    interest_df = interest_df.fillna(0)
    # Create an expected column with the proportion of non-matches per interest level difference
    interest_df['expected'] = interest_df['unmatched']/total_unmatched
    # Create an observed column with the proportion of matches per interest level difference
    interest_df['observed'] = interest_df['matched']/total_matched
    # Find the p-value based on the chi-squared test
    pval = stats.chisquare(interest_df['observed'], interest_df['expected']).pvalue
    if pval > 0.05:
        print(f'According to the chi-square test, the p-value for {activity} was {pval}.')
        print(f'Since the p-value is greater than 0.05, we fail to reject the null hypothesis')
        print(f'because the difference between the observed and expected values is not statistically significant.')
        print(f'In other words, {activity} has no influence on more or less matches.')
        print(f'i.e. it does not influence if the speed date will result in a match.')
    else:
        print(f'According to the chi-square test, the p-value for {activity} was {pval}.')
        print(f'Since the p-value is less than 0.05, we reject the null hypothesis')
        print(f'because the difference between the observed and expected values is statistically significant.')
        print(f'In other words, {activity} has some influence on more or less matches.')
        print(f'i.e. it may influence if the speed date will result in a match.')
```

Chi-squared tests were used to identity the goodness of fit between the observed population (i.e. matches) and the expected population (i.e. non-matches). The null hypothesis is that there is no association between the observed and the expected population. In order words, the null hypothesis is that there is no influence on whether or not a couple will match based on their shared or opposing level of interest in an activity.

Based on the chi-square tests above, all of them failed to reject the null hypothesis. In other words, difference in interest levels among the 17 activities has no influence on whether or not a couple will be attracted to each other i.e. that the pair will result in a match.

Therefore, based on the findings above, shared interest levels hold no conclusion on whether opposites attract. While there was no strong correlation on shared interests in the heatmaps, the absolute difference showed that matches occured more frequently when a couple had similar interest levels in an activity (i.e absolute difference less than 3). However, after completing statistical analysis using chi-squared tests, we failed to reject the null that there is no association between matches and non-matches based on interest.

In the dataset, individuals were asked, "what was their primary goal in participating in the speed dating event?" Below is the provided classification based on the data key.

goal:

This classification was then further categorized based on whether the participant was looking for something serious (i.e. answered either 3 or 4) or casual (i.e. any other responses besides 3 or 4). This categorization was performed in the data exploration and clean up sections. In addition, anyone who answered 'Other' (6) was removed from the data since the answer was too ambiguous to categorize.

The 'goal' column was utilized to measure a participant's intention from the speed dating event and to investigate if two people looking for a relationship or two people looking for something casual will match less frequently than those with opposing dating intensions.

```python
# Create a copy of the combo_goal in order to manipulate the dataframe
counts_goal = combo_goal.copy()
# Use groupby to separate data into fields based on individual and partner goal regardless if they matched or not
counts_goal = counts_goal.groupby(["seriousness_ind", "seriousness_par"]).size().reset_index(name="times_met")

# Categorize each pair with s=serious and c=casual
ss = int(counts_goal.loc[(counts_goal.seriousness_ind == 'serious') & (counts_goal.seriousness_par == 'serious'), 'times_met'])
cc = int(counts_goal.loc[(counts_goal.seriousness_ind == 'casual') & (counts_goal.seriousness_par == 'casual'), 'times_met'])
cs = int(counts_goal.loc[(counts_goal.seriousness_ind == 'casual') & (counts_goal.seriousness_par == 'serious'), 'times_met'])
sc = int(counts_goal.loc[(counts_goal.seriousness_ind == 'serious') & (counts_goal.seriousness_par == 'casual'), 'times_met'])
# Combine the two serious-casual groups into one
sc = cs + sc

# Create a bar plot to show the three categorization of intention
plt.bar(['Both Serious', 'Both Casual', 'Serious-Casual'],[ss, cc, sc])
# Label the y-axis accordingly
plt.ylabel('Number of Speed Dates')
# Set an appropriate title
plt.title("Number of Speed Dates Based on Couple's Intention",loc='center')
# Save as .png
plt.savefig("Visuals/intention/Number of Speed Dates Based on Couple's Intention.png", transparent=False)
# Display the plot
plt.show()
```
